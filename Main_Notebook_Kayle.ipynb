# Kyle Model: discrete model

In the **Kyle model**, we have three main actors:

1. **Insider**: Knows the value of the asset $P$ and aims to maximize their profit by choosing the **order volume $V$** to place in the market.
2. **Noise Traders**: $n$ represents their order volumes.
3. **Market Makers**: Observe the total demand $Y = V + n$ and set a market price $h(Y)$ to minimize their losses.

---

## Notations

- $P$: Asset value (random variable).
- $n$: Orders from **noise traders** (random variable).
- $V$: Insider's order volume (variable to be determined).
- $Y$: Total demand, given by $Y = V + n$.
- $h(Y)$: Market price set by the market maker (variable to be determined).
- For a possible outcome $\omega$, $P_\omega$ and $n_\omega$ denote specific realizations of $P$ and $n$.

---

## Insider's Optimization Problem

The insider knows the asset value $P$ at $t = 0$ and aims to maximize their **profit** by determining $V$. The market price set by the market makers depends on the observed total demand $Y = V + n$.

### Insider's Profit

For an outcome $\omega$, the insider's profit is:

$
\text{Profit}_\text{insider} = (P_\omega - h(Y_\omega)) \cdot V
$

### Insider's Maximization Problem

The insider seeks to maximize their profit by choosing $V$:

$
\max_V \mathbb{E}\left[ (P - h(V + n)) \cdot V \right]
$

---

## Market Maker's Optimization Problem

The market makers aim to set $h(Y)$ such that their expected profit is zero. The profit for a market maker is given by:

### Market Makers' Profit

$
\text{Profit}_\text{MM} = (h(Y) - P) \cdot Y
$

The market maker satisfies the condition:

$
\mathbb{E}\left[ (h(Y) - P) \cdot Y \right] = 0
$

**Note**: In the Kyle model, market makers are perfectly competitive (Bertrand competition) and risk-neutral. They set the price such that their expected profit is zero, implying no profit or loss on average.

---

## Assumptions and Notations

- $P$: Real value of the asset, modeled as a random variable.
- $V$: Insider's order volume, proportional to the asset value, $V = \beta P$.
- $n$: Noise traders' order volume, modeled as a random variable independent of $P$.
- $Y$: Total demand observed by market makers, $Y = V + n = \beta P + n$.
- $h(Y)$: Market price set by market makers, depending on the total demand $Y$.

We assume:

- $P \sim \mathcal{N}(0, \sigma_P^2)$
- $n \sim \mathcal{N}(0, \sigma_n^2)$

Additionally, we assume:

1. Insider's order volume is linear in $P$:

$
V = \beta \cdot P + b
$

2. Market price set by market makers is linear in $Y$:

$
h(Y) = \lambda \cdot Y + a
$

---

## Theoretical Results for Optimal Parameters

The optimal parameters are:

$
a = b = 0, \quad \beta = \frac{\sigma_n}{\sigma_P}, \quad \lambda = \frac{\sigma_P}{2\sigma_n}
$


# Solving the Linear Kyle Model

---

### Assumptions Recap:

1. $ P \sim \mathcal{N}(0, \sigma_P^2) $
2. $ n \sim \mathcal{N}(0, \sigma_n^2) $, independent of $P$.
3. Insider's order volume:
   $
   V = \beta P + b
   $
4. Market price:
   $
   h(Y) = \lambda Y + a
   $
   where $ Y = V + n = \beta P + n $.

---

### 1. **Market Makers' Zero-Expected-Profit Condition**

Market makers set $h(Y)$ to ensure:
$
\mathbb{E} \left[ (h(Y) - P) \cdot Y \right] = 0.
$

Substitute $ h(Y) = \lambda Y + a $ and $ Y = \beta P + n $:
$
\mathbb{E} \left[ (\lambda Y + a - P) \cdot Y \right] = 0.
$

Expand:
$
\mathbb{E} \left[ \lambda Y^2 + aY - PY \right] = 0.
$

#### Calculations:

1. $ \mathbb{E}[Y^2] $:
   $
   Y = \beta P + n, \quad Y^2 = \beta^2 P^2 + 2\beta P n + n^2.
   $
   Since $P$ and $n$ are independent:
   $
   \mathbb{E}[Y^2] = \beta^2 \mathbb{E}[P^2] + \mathbb{E}[n^2] = \beta^2 \sigma_P^2 + \sigma_n^2.
   $

2. $ \mathbb{E}[PY] $:
   $
   PY = P(\beta P + n) = \beta P^2 + Pn.
   $
   Again, independence gives:
   $
   \mathbb{E}[PY] = \beta \mathbb{E}[P^2] = \beta \sigma_P^2.
   $

3. $ \mathbb{E}[Y] $:
   $
   \mathbb{E}[Y] = \beta \mathbb{E}[P] + \mathbb{E}[n] = 0.
   $

Substituting into the zero-profit condition:
$
\lambda (\beta^2 \sigma_P^2 + \sigma_n^2) + a \cdot 0 - \beta \sigma_P^2 = 0.
$

Solve for $\lambda$:
$
\lambda = \frac{\beta \sigma_P^2}{\beta^2 \sigma_P^2 + \sigma_n^2}.
$

---

### 2. **Insider's Profit Maximization**

The insider maximizes:
$
\mathbb{E}[(P - h(Y)) \cdot V].
$

Substitute $V = \beta P + b$ and $h(Y) = \lambda Y + a$:
$
\mathbb{E}[(P - (\lambda Y + a)) \cdot (\beta P + b)].
$

Expand:
$
\mathbb{E}[\beta P^2 - \beta \lambda PY - \beta a P + bP - b\lambda Y - ba].
$

#### Calculations:

1. $ \mathbb{E}[P^2] $:
   $
   \mathbb{E}[P^2] = \sigma_P^2.
   $

2. $ \mathbb{E}[PY] $:
   $
   \mathbb{E}[PY] = \beta \sigma_P^2.
   $

3. $ \mathbb{E}[Y] $:
   $
   \mathbb{E}[Y] = 0.
   $

The profit simplifies to:
$
\beta \sigma_P^2 - \beta \lambda (\beta \sigma_P^2).
$

Maximize with respect to $\beta$:
$
\frac{\partial}{\partial \beta} \left[ \beta \sigma_P^2 (1 - \lambda \beta) \right] = 0.
$

Solve:
$
\sigma_P^2 (1 - 2\lambda \beta) = 0 \quad \implies \quad \beta = \frac{1}{2\lambda}.
$

---

### Final Parameters

Using the expressions for $\lambda$ and substituting:
1. $\beta = \frac{\sigma_n}{\sigma_P}$,
2. $\lambda = \frac{\sigma_P}{2\sigma_n}$,


$
\beta = \frac{1}{2\lambda}
$

# Why $a = 0$ and $b = 0$ in the Kyle Model?

In the linear Kyle model, the constants $a$ and $b$ can theoretically take any value, but they are typically set to $0$ for simplicity and clarity. Here's why:

---

### 1. $a = 0$: Market Price Offset
The market price is given by:
$
h(Y) = \lambda Y + a.
$
- $a$ acts as an arbitrary offset in the price.
- Under the assumptions of the model ($\mathbb{E}[P] = 0$, $\mathbb{E}[n] = 0$), the market price is naturally unbiased. Setting $a = 0$ ensures:
$
\mathbb{E}[h(Y)] = \mathbb{E}[P].
$
- Any non-zero $a$ would only shift the price without affecting equilibrium conditions, making $a = 0$ the simplest choice.

---

### 2. $b = 0$: Insider's Order Volume
The insider's order volume is:
$
V = \beta P + b.
$
- The constant $b$ does not depend on $P$ (the insider's informational advantage). Hence, it does not contribute to maximizing profit:
$
\text{Profit}_\text{insider} = \mathbb{E} \left[ (P - h(V + n)) \cdot (\beta P + b) \right].
$
- Introducing $b$ only adds a redundant, non-strategic term.

---

### 3. Simplicity and Normalization
- Setting $a = 0$ and $b = 0$ simplifies the model while preserving its core dynamics.
- The equilibrium results remain unaffected because $a$ and $b$ do not alter the strategic interactions between the insider and market makers.

---

### Conclusion
While $a$ and $b$ can theoretically take any value, setting them to zero:
- Simplifies calculations and interpretations.
- Reflects the unbiased nature of market makers and the pure strategic role of the insider.


# Numerical Resolution of the Linear Kyle Model

---

### 1. Linear Model Assumptions

The linear model assumes:
$
V = \beta P, \quad h(Y) = \lambda Y, \quad Y = \beta P + n,
$
where:
$
P \sim \mathcal{N}(0, \sigma_P^2), \quad n \sim \mathcal{N}(0, \sigma_n^2).
$

---

### 2. Steps for Monte Carlo Simulation

#### **Step 1: Generate Random Variables**
Generate samples of $P$ and $n$ from their respective normal distributions:
$
P \sim \mathcal{N}(0, \sigma_P^2), \quad n \sim \mathcal{N}(0, \sigma_n^2).
$

#### **Step 2: Iteratively Optimize for $\beta$**
For each value of $\lambda$:
1. Solve the insider's profit maximization problem:
   $
   \max_\beta \mathbb{E} \left[ (P - h(Y)) \cdot V \right] = \max_\beta \mathbb{E} \left[ (P - \lambda (\beta P + n)) \cdot (\beta P) \right].
   $
2. Include the market maker's zero-expected-profit constraint:
   $
   \mathbb{E} \left[ (h(Y) - P) \cdot Y \right] = 0.
   $
3. Numerically optimize $\beta$ for each $\lambda$.

#### **Step 3: Plot $\beta(\lambda)$**
- Store the results for $\beta$ as a function of $\lambda$.
- Plot $\beta$ vs $\lambda$.

#### **Step 4: Solve for Optimal $\lambda$**
- Substitute $\beta(\lambda)$ into the insider's profit maximization problem:
  $
  \max_\lambda \mathbb{E} \left[ (P - \lambda (\beta(\lambda) P + n)) \cdot (\beta(\lambda) P) \right].
  $
- Solve for $\lambda$ numerically.

#### **Step 5: Deduce $\beta$**
Using the optimal $\lambda$, compute the corresponding $\beta$.

---

### Outcome
- A plot of $\beta(\lambda)$.
- The optimal values of $\lambda$ and $\beta$.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Parameters
sigma_P = 1.0  # Variance of P
sigma_n = 1.0  # Variance of n
num_samples = 10000  # Number of Monte Carlo samples

# Generate Monte Carlo samples
P = np.random.normal(0, sigma_P, num_samples)
n = np.random.normal(0, sigma_n, num_samples)

# Theoretical values for comparison
theoretical_beta = sigma_n / sigma_P
theoretical_lambda = sigma_P / (2 * sigma_n)

# Insider profit function
def insider_profit(beta, lambda_, P, n):
    Y = beta * P + n
    h_Y = lambda_ * Y
    return -np.mean((P - h_Y) * (beta * P))  # Negative for minimization

# Market maker constraint function
def market_maker_constraint(beta, lambda_, P, n):
    Y = beta * P + n
    h_Y = lambda_ * Y
    return np.mean((h_Y - P) * Y)

# Find beta for each lambda
lambda_values = np.linspace(0.1, 5, 1000)
beta_values = []

for lambda_ in lambda_values:
    # Objective function for insider profit maximization
    def objective(beta):
        return insider_profit(beta, lambda_, P, n)
    
    # Constraint for market makers
    constraint = {
        'type': 'eq',
        'fun': lambda beta: market_maker_constraint(beta, lambda_, P, n)
    }
    
    # Solve for optimal beta
    result = minimize(objective, x0=0.5, constraints=constraint, bounds=[(0, None)])
    beta_values.append(result.x[0])

# Plot beta vs lambda
plt.figure(figsize=(8, 6))
plt.plot(lambda_values, beta_values, label=r'Numerical $\beta(\lambda)$')
plt.axhline(y=theoretical_beta, color='r', linestyle='--', label=r'Theoretical $\beta = \frac{\sigma_n}{\sigma_P}$')
plt.xlabel(r'$\lambda$')
plt.ylabel(r'$\beta$')
plt.title(r'Optimal $\beta$ as a function of $\lambda$')
plt.legend()
plt.grid()
plt.show()

# Solve for optimal lambda
def reduced_objective(lambda_):
    # Interpolate beta(lambda) and compute insider profit
    beta_lambda = np.interp(lambda_, lambda_values, beta_values)
    return insider_profit(beta_lambda, lambda_, P, n)

result_lambda = minimize(reduced_objective, x0=1.0, bounds=[(0.1, 5)])
optimal_lambda = result_lambda.x[0]
optimal_beta = np.interp(optimal_lambda, lambda_values, beta_values)

# Print results
print(f"Theoretical values:")
print(f"  Beta (theoretical): {theoretical_beta:.4f}")
print(f"  Lambda (theoretical): {theoretical_lambda:.4f}\n")

print(f"Numerical results:")
print(f"  Optimal lambda (numerical): {optimal_lambda:.4f}")
print(f"  Optimal beta (numerical): {optimal_beta:.4f}")
```

# Reversing the Approach: Solving for $\lambda$

Instead of solving for $\beta$ for fixed $\lambda$, we now:

1. **Fix $\beta$**: Choose a range of values for $\beta$.
2. **Solve for $\lambda$**: For each fixed $\beta$, find the optimal $\lambda$ that:
   - Maximizes the insider's profit:
     $
     \max_\lambda \mathbb{E}\left[(P - \lambda (\beta P + n)) \cdot (\beta P)\right]
     $
   - Satisfies the market maker's zero-expected-profit condition:
     $
     \mathbb{E}\left[(\lambda (\beta P + n) - P) \cdot (\beta P + n)\right] = 0.
     $
3. **Plot $\lambda(\beta)$**: Visualize the relationship between $\lambda$ and $\beta$.
4. **Reinject $\lambda(\beta)$**: Use the function $\lambda(\beta)$ to simplify the insider's profit optimization problem to a single-variable problem in $\beta$:
   $
   \max_\beta \mathbb{E}\left[(P - \lambda(\beta)(\beta P + n)) \cdot (\beta P)\right].
   $
5. **Solve for Optimal $\beta$**: Find the $\beta$ that maximizes the insider's profit.

This approach mirrors the first but focuses on deriving $\lambda$ for fixed $\beta$, allowing us to analyze the inverse relationship between the parameters.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize

# Parameters
sigma_P = 1.0  # Variance of P
sigma_n = 1.0  # Variance of n
num_samples = 10000  # Number of Monte Carlo samples

# Generate Monte Carlo samples
P = np.random.normal(0, sigma_P, num_samples)
n = np.random.normal(0, sigma_n, num_samples)

# Theoretical values for comparison
theoretical_beta = sigma_n / sigma_P
theoretical_lambda = sigma_P / (2 * sigma_n)

# Insider profit function
def insider_profit(beta, lambda_, P, n):
    Y = beta * P + n
    h_Y = lambda_ * Y
    return -np.mean((P - h_Y) * (beta * P))  # Negative for minimization

# Market maker constraint function
def market_maker_constraint(beta, lambda_, P, n):
    Y = beta * P + n
    h_Y = lambda_ * Y
    return np.mean((h_Y - P) * Y)

# Find lambda for each beta
beta_values = np.linspace(0.1, 5, 50)
lambda_values = []

for beta in beta_values:
    # Objective function for insider profit maximization
    def objective(lambda_):
        return insider_profit(beta, lambda_, P, n)
    
    # Constraint for market makers
    constraint = {
        'type': 'eq',
        'fun': lambda lambda_: market_maker_constraint(beta, lambda_, P, n)
    }
    
    # Solve for optimal lambda
    result = minimize(objective, x0=1.0, constraints=constraint, bounds=[(0, None)])
    lambda_values.append(result.x[0])

# Plot lambda vs beta
plt.figure(figsize=(8, 6))
plt.plot(beta_values, lambda_values, label=r'Numerical $\lambda(\beta)$')
plt.axhline(y=theoretical_lambda, color='r', linestyle='--', label=r'Theoretical $\lambda = \frac{\sigma_P}{2\sigma_n}$')
plt.xlabel(r'$\beta$')
plt.ylabel(r'$\lambda$')
plt.title(r'Optimal $\lambda$ as a function of $\beta$')
plt.legend()
plt.grid()
plt.show()

# Solve for optimal beta
def reduced_objective(beta):
    # Interpolate lambda(beta) and compute insider profit
    lambda_beta = np.interp(beta, beta_values, lambda_values)
    return insider_profit(beta, lambda_beta, P, n)

result_beta = minimize(reduced_objective, x0=1.0, bounds=[(0.1, 5)])
optimal_beta = result_beta.x[0]
optimal_lambda = np.interp(optimal_beta, beta_values, lambda_values)

# Print results
print(f"Theoretical values:")
print(f"  Beta (theoretical): {theoretical_beta:.4f}")
print(f"  Lambda (theoretical): {theoretical_lambda:.4f}\n")

print(f"Numerical results:")
print(f"  Optimal beta (numerical): {optimal_beta:.4f}")
print(f"  Optimal lambda (numerical): {optimal_lambda:.4f}")
```

---

### 1. Observations

- The second approach, where we solve for $\lambda(\beta)$, produces results closer to the theoretical values than the first method ($\beta(\lambda)$).
- The function $\lambda(\beta)$ appears to be **smoother** and well-behaved, indicating that it is a $C^\infty$ (infinitely differentiable) function over the domain of $\beta$.
- Conversely, the function $\beta(\lambda)$ shows instability, particularly as $\lambda$ deviates from the theoretical equilibrium point.

---

### 2. Analysis of $\lambda(\beta)$

The function $\lambda(\beta)$:
- Represents the market makers’ optimal response to the insider’s strategy ($\beta$).
- Aligns naturally with the market makers’ zero-expected-profit constraint:
  $
  \mathbb{E}\left[ (\lambda (\beta P + n) - P) \cdot (\beta P + n) \right] = 0.
  $
- Smoothly adapts to changes in $\beta$, making the relationship between $\lambda$ and $\beta$ more stable.

---

### 3. Analysis of $\beta(\lambda)$

The function $\beta(\lambda)$:
- Solves for the insider’s optimal strategy while satisfying the market makers’ pricing:
  $
  \max_\beta \mathbb{E}\left[ (P - \lambda (\beta P + n)) \cdot (\beta P) \right].
  $
- Is more sensitive to variations in $\lambda$, as small deviations in $\lambda$ can significantly affect the insider’s profit.
- This sensitivity introduces numerical instabilities in the optimization process, particularly when $\lambda$ is far from equilibrium.

---

### 4. Comparison and Key Insights

- The function $\lambda(\beta)$ is well-suited for optimization due to its smoothness ($C^\infty$) and stability.
- Solving for $\lambda(\beta)$ first and reinjecting it into the insider’s profit maximization problem reduces the optimization to a single variable ($\beta$), leading to better numerical results.
- The theoretical values ($\lambda = \frac{\sigma_P}{2\sigma_n}$, $\beta = \frac{\sigma_n}{\sigma_P}$) serve as excellent benchmarks, and $\lambda(\beta)$ converges more closely to these theoretical predictions.

---

### 5. Conclusion

- Solving for $\lambda(\beta)$ is preferred due to its smooth behavior and closer alignment with theoretical expectations.
- This method not only improves numerical stability but also provides better insights into the equilibrium dynamics of the Kyle model.
- Future work can focus on further analyzing the regularity of $\lambda(\beta)$ and exploring its analytical properties to enhance understanding of the model.

# Function to Solve the Kyle Model

We define a function to solve the Kyle model using the reinjection method. This function:
1. Calculates the optimal insider strategy ($\beta$) and market maker sensitivity ($\lambda$).
2. Incorporates both the insider's profit maximization and the market maker's zero-profit constraint.
3. Leverages numerical optimization to dynamically compute $\lambda(\beta)$ and subsequently optimize $\beta$.

The function is designed to handle any distribution of $P$ (asset value) and $n$ (noise traders), enabling flexible analysis across various market conditions.

```python
import numpy as np
from scipy.optimize import minimize

def solve_kyle_model(P, n, beta_range=None, lambda_range_factor=0.5, num_points=1000):
    """
    Solves for the optimal beta and lambda using the reinjection method.

    Parameters:
        P (np.ndarray): Array of sampled values of P (asset value).
        n (np.ndarray): Array of sampled values of n (noise traders).
        beta_range (tuple, optional): Range of beta values (min, max). Defaults to None.
        lambda_range_factor (float, optional): Factor to set the range of lambda values
                                               around the theoretical value.
        num_points (int): Number of points to sample for beta and lambda.

    Returns:
        dict: A dictionary containing the optimal beta, optimal lambda, and a grid of beta/lambda values.
    """
    # Calculate standard deviations from the samples
    sigma_P = np.std(P)
    sigma_n = np.std(n)

    # Theoretical values for comparison
    theoretical_beta = sigma_n / sigma_P
    theoretical_lambda = sigma_P / (2 * sigma_n)

    # Define ranges for beta and lambda
    if beta_range is None:
        beta_range = (0.1, 5)  # Default range for beta
    lambda_range = (theoretical_lambda * (1 - lambda_range_factor), 
                    theoretical_lambda * (1 + lambda_range_factor))

    # Create ranges for beta and lambda
    beta_values = np.linspace(beta_range[0], beta_range[1], num_points)
    lambda_values = np.linspace(lambda_range[0], lambda_range[1], num_points)

    # Insider profit function (gain)
    def insider_gain(beta, lambda_):
        Y = beta * P + n
        h_Y = lambda_ * Y
        return np.mean((P - h_Y) * (beta * P))

    # Market maker constraint function
    def market_maker_constraint(beta, lambda_):
        Y = beta * P + n
        h_Y = lambda_ * Y
        return np.mean((h_Y - P) * Y)

    # Solve for lambda as a function of beta
    optimal_lambdas = []
    for beta in beta_values:
        def objective(lambda_):
            return -insider_gain(beta, lambda_)  # Maximize insider gain (minimize negative)

        constraint = {
            'type': 'eq',
            'fun': lambda lambda_: market_maker_constraint(beta, lambda_)
        }

        result = minimize(objective, x0=theoretical_lambda, constraints=constraint, bounds=[(0, None)])
        optimal_lambdas.append(result.x[0])

    # Interpolation function for lambda(beta)
    def lambda_of_beta(beta):
        return np.interp(beta, beta_values, optimal_lambdas)

    # Reduce to single-variable optimization problem for beta
    def reduced_objective(beta):
        lambda_beta = lambda_of_beta(beta)
        return -insider_gain(beta, lambda_beta)  # Maximize insider gain

    result_beta = minimize(reduced_objective, x0=theoretical_beta, bounds=[(beta_range[0], beta_range[1])])
    optimal_beta = result_beta.x[0]
    optimal_lambda = lambda_of_beta(optimal_beta)

    # Return results
    return {
        "optimal_beta": optimal_beta,
        "optimal_lambda": optimal_lambda,
        "beta_values": beta_values,
        "lambda_values": lambda_values,
        "optimal_lambdas": optimal_lambdas,
        "theoretical_beta": theoretical_beta,
        "theoretical_lambda": theoretical_lambda,
    }
```

```python
# Example usage:
# Generate sample data for P and n
np.random.seed(42)
P_sample = np.random.normal(0, 1, 10000)  # Asset values
n_sample = np.random.normal(0, 1, 10000)  # Noise trader values

# Solve for optimal beta and lambda
results = solve_kyle_model(P_sample, n_sample)

# Print results
print("Optimal Beta:", results["optimal_beta"])
print("Optimal Lambda:", results["optimal_lambda"])
print("Theoretical Beta:", results["theoretical_beta"])
print("Theoretical Lambda:", results["theoretical_lambda"])
```

# Solving the Kyle Model Using TensorFlow

### 1. **Dynamic Optimization Framework**

TensorFlow provides a gradient-based optimization framework that enables efficient computation of derivatives and optimization over multi-dimensional spaces. This is crucial for solving the Kyle model, which involves:
- **Maximizing insider gain**:
  $
  \text{Gain}_{\text{insider}} = \mathbb{E}\left[(P - h(Y)) \cdot (\beta P)\right]
  $
- **Satisfying the market maker zero-profit constraint**:
  $
  \mathbb{E}\left[(h(Y) - P) \cdot Y\right] = 0
  $

Here:
- $P$: Asset value.
- $n$: Noise trader volumes.
- $Y = \beta P + n$: Total observed demand.
- $h(Y) = \lambda Y$: Market price set by market makers.

### 2. **Problem Setup**

#### Variables:
- $\beta$: Insider's order volume strategy (optimized).
- $\lambda$: Market maker's price sensitivity (optimized).

#### Loss Function:
A combined loss function is defined to incorporate both objectives:
$
\text{Loss} = -\text{Gain}_{\text{insider}} + \text{Penalty}_{\text{constraint}}
$
where:
$
\text{Penalty}_{\text{constraint}} = \left|\mathbb{E}\left[(h(Y) - P) \cdot Y\right]\right|
$

#### Data:
- Random samples of $P$ and $n$ are generated to simulate the market environment. These samples can follow various distributions.

---

## TensorFlow Implementation

### 1. **Gradient-Based Optimization**
TensorFlow computes gradients automatically for all operations involving trainable variables ($\beta$ and $\lambda$). The optimizer (Adam) updates the variables to minimize the loss function.

### 2. **Computation Graph**
TensorFlow builds a computation graph dynamically, where:
- $\beta$ and $\lambda$ are trainable variables.
- The demand function $Y$ and market price $h(Y)$ are computed dynamically.
- Gradients are propagated through the graph to update $\beta$ and $\lambda$.

```python
import tensorflow as tf
import numpy as np

# Generate synthetic data for P (asset value) and n (noise traders)
np.random.seed(42)
num_samples = 10000
sigma_P, sigma_n = 1.0, 1.0
P = np.random.normal(0, sigma_P, num_samples)
n = np.random.normal(0, sigma_n, num_samples)

# Theoretical values for reference
theoretical_beta = sigma_n / sigma_P
theoretical_lambda = sigma_P / (2 * sigma_n)

# Convert data to TensorFlow tensors
P_tf = tf.constant(P, dtype=tf.float32)
n_tf = tf.constant(n, dtype=tf.float32)

# Define TensorFlow variables with random initialization
beta = tf.Variable(initial_value=np.random.uniform(0.1, 5), dtype=tf.float32, name="beta", trainable=True)
lambda_ = tf.Variable(initial_value=np.random.uniform(0.1, 5), dtype=tf.float32, name="lambda", trainable=True)

# Define the demand function Y
def calculate_Y(beta, lambda_):
    Y_tf = beta * P_tf + n_tf
    return Y_tf

# Insider gain (negative for maximization in TensorFlow)
def insider_gain(beta, lambda_):
    Y_tf = calculate_Y(beta, lambda_)
    h_Y = lambda_ * Y_tf
    return tf.reduce_mean((P_tf - h_Y) * (beta * P_tf))

# Market maker zero-profit constraint
def market_maker_constraint(beta, lambda_):
    Y_tf = calculate_Y(beta, lambda_)
    h_Y = lambda_ * Y_tf
    return tf.reduce_mean((h_Y - P_tf) * Y_tf)

# Define a combined loss function
def loss():
    gain = insider_gain(beta, lambda_)
    constraint_penalty = tf.abs(market_maker_constraint(beta, lambda_))
    return -gain + constraint_penalty  # Maximize insider gain while satisfying constraints

# Optimizer
optimizer = tf.keras.optimizers.Adam(learning_rate=0.01)

# Optimization loop
epochs = 1000
for epoch in range(epochs):
    with tf.GradientTape() as tape:
        current_loss = loss()
    gradients = tape.gradient(current_loss, [beta, lambda_])
    optimizer.apply_gradients(zip(gradients, [beta, lambda_]))

    if epoch % 100 == 0:
        print(f"Epoch {epoch}, Loss: {current_loss.numpy()}, Beta: {beta.numpy()}, Lambda: {lambda_.numpy()}")

# Final values
optimal_beta = beta.numpy()
optimal_lambda = lambda_.numpy()

print("\nOptimal Beta:", optimal_beta)
print("Optimal Lambda:", optimal_lambda)
print("Theoretical Beta:", theoretical_beta)
print("Theoretical Lambda:", theoretical_lambda)
```

# Simulating Price Action Using the Kyle Model with Dynamic Optimization

---

## 1. **Objective**

The goal is to simulate how market prices evolve over time using the **Kyle model**, incorporating dynamic optimization to calculate the insider’s strategy ($\beta$) and the market makers’ sensitivity to order flow ($\lambda$). This refined approach addresses periods of market **consolidation** and **information shocks**, where the volatilities of the asset ($\sigma_P$) and noise traders ($\sigma_n$) vary dynamically.

---

## 2. **Model Assumptions**

We assume the following:

- **Asset Value ($P$)**: Modeled as a random variable with time-dependent volatility $\sigma_P$:
  $
  P \sim \mathcal{N}(0, \sigma_P^2)
  $
- **Noise Trader Volume ($n$)**: Modeled as a random variable independent of $P$ with time-dependent volatility $\sigma_n$:
  $
  n \sim \mathcal{N}(0, \sigma_n^2)
  $
- **Insider’s Order Volume ($V$)**: Linear in $P$:
  $
  V = \beta P
  $
- **Total Demand ($Y$)**: Observed by market makers:
  $
  Y = V + n = \beta P + n
  $
- **Market Price ($h(Y)$)**: Linear in $Y$:
  $
  h(Y) = \lambda Y
  $

### Theoretical Parameters

The Kyle model predicts the following theoretical values:
$
\beta = \frac{\sigma_n}{\sigma_P}, \quad \lambda = \frac{\sigma_P}{2 \sigma_n}
$

---

## 3. **Dynamic Simulation Setup**

### 3.1 **Dynamic Volatility ($\sigma_P$ and $\sigma_n$)**
- **Consolidation Periods**: High $\sigma_n$ (dominance of noise traders) and low $\sigma_P$ (low insider activity).
- **Information Shock**: Sudden increase in $\sigma_P$ (insider reacts to new information) and decrease in $\sigma_n$.

At every time step $t$, $\sigma_P$ and $\sigma_n$ are updated based on the current regime:
1. **Consolidation**: $\sigma_P \ll \sigma_n$
2. **Shock**: $\sigma_P \gg \sigma_n$

---

### 3.2 **Sliding Window for Market Makers**
Market makers use a **sliding window** approach to incorporate recent price action:
- Window duration: 4 hours.
- At each time step $t$, market makers observe $(P, n)$ over the last 4 hours to estimate $\lambda$.

---

### 3.3 **Optimization for $\lambda(\beta)$**
For a fixed insider strategy ($\beta$):
1. **Insider's Profit Maximization**:
   $
   \max_{\lambda} \mathbb{E}\left[(P - h(Y)) \cdot (\beta P)\right]
   $
2. **Market Maker's Zero-Profit Constraint**:
   $
   \mathbb{E}\left[(h(Y) - P) \cdot Y\right] = 0
   $

$\lambda(\beta)$ is solved dynamically at every time step using the sliding window.

---

### 3.4 **Dynamic Insider Strategy ($\beta$)**
The insider’s strategy ($\beta$) is calculated dynamically as:
$
\beta = \frac{\sigma_n}{\sigma_P}
$
This ensures the insider adapts their trading volume to the relative volatilities of the asset and noise traders.

---

## 4. **Simulation Workflow**

1. **Initialize Parameters**:
   - Define total simulation time and time step ($\Delta t$).
   - Initialize volatilities $\sigma_P$ and $\sigma_n$.

2. **Generate Samples**:
   - Generate random samples for $P$ and $n$ using the respective volatilities.

3. **Sliding Window Optimization**:
   - At each time step $t$, use the last 4 hours of data to:
     1. Solve for $\lambda(\beta)$ using dynamic optimization.
     2. Calculate the market price $h(Y)$ and insider’s profit.

4. **Update Cumulative Values**:
   - Track cumulative market prices, true asset values, and insider profits over time.

5. **Dynamic Regime Shifts**:
   - Alternate between consolidation and shock periods by updating $\sigma_P$ and $\sigma_n$ every 3 hours.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from collections import namedtuple

# Define the TimeSlice structure
TimeSlice = namedtuple('TimeSlice', ['t', 'P', 'n'])

# Parameters for the simulation
total_hours = 24     # Total duration in hours
delta_t = 0.1        # Time step in hours (6 minutes)
times = np.arange(0, total_hours + delta_t, delta_t)  # Time array

# Duration of each regime (consolidation or shock)
interval_length = 3  # 3 hours
interval_points = int(interval_length / delta_t)

# Generate dynamic sigma_P and sigma_n values
sigma_P_values = []
sigma_n_values = []
interval_boundaries = []  # To store the interval boundaries for plotting

for i, t in enumerate(times):
    # Mark the interval boundaries
    if i % interval_points == 0:
        interval_boundaries.append(t)
    
    if 12 <= t < 15:  # Shock period
        sigma_P = 2.0
        sigma_n = 0.5
    else:  # Consolidation period
        sigma_P = np.random.uniform(0.1, 0.5)
        sigma_n = np.random.uniform(1.5, 3.0)
    sigma_P_values.append(sigma_P)
    sigma_n_values.append(sigma_n)

# Generate P and n values based on sigma_P and sigma_n
P_values = np.random.normal(0, np.array(sigma_P_values) * np.sqrt(delta_t))
n_values = np.random.normal(0, np.array(sigma_n_values) * np.sqrt(delta_t))

# Function to solve for optimal lambda given beta
def solve_lambda(beta, P_window, n_window):
    def objective(lambda_):
        Y_window = beta * P_window + n_window
        h_Y = lambda_ * Y_window
        return -np.mean((P_window - h_Y) * (beta * P_window))  # Insider gain (negative for maximization)
    
    def constraint(lambda_):
        Y_window = beta * P_window + n_window
        h_Y = lambda_ * Y_window
        return np.mean((h_Y - P_window) * Y_window)  # Market maker zero-profit constraint
    
    result = minimize(objective, x0=0.1, constraints={'type': 'eq', 'fun': constraint}, bounds=[(0, None)])
    return result.x[0] if result.success else 0.0

# Sliding window simulation
window_duration = 4  # 4 hours
window_size = int(window_duration / delta_t)

lambda_values = []
h_Y_values = []
insider_profits = []
market_prices_cumulative = []
true_values_cumulative = []

for i in range(len(times)):
    # Define the sliding window
    start_idx = max(0, i - window_size + 1)
    P_window = P_values[start_idx:i+1]
    n_window = n_values[start_idx:i+1]
    
    # Solve for beta and lambda dynamically
    sigma_P = sigma_P_values[i]
    sigma_n = sigma_n_values[i]
    beta = sigma_n / sigma_P  # Theoretical beta
    lambda_mm = solve_lambda(beta, P_window, n_window)
    
    # Calculate V, Y, and h(Y) for the current time step
    V_current = beta * P_values[i]
    Y_current = V_current + n_values[i]
    h_Y_current = lambda_mm * Y_current
    
    # Calculate profits
    insider_profit = (P_values[i] - h_Y_current) * V_current
    insider_profits.append(insider_profit)
    
    # Track cumulative values
    if i == 0:
        market_prices_cumulative.append(h_Y_current)
        true_values_cumulative.append(P_values[i])
    else:
        market_prices_cumulative.append(market_prices_cumulative[-1] + h_Y_current)
        true_values_cumulative.append(true_values_cumulative[-1] + P_values[i])
    
    lambda_values.append(lambda_mm)

# Plot cumulative market prices vs true values
plt.figure(figsize=(12, 6))
plt.plot(times, market_prices_cumulative, label='Cumulative Market Price h(Y)', color='blue')
plt.plot(times, true_values_cumulative, label='Cumulative True Value P', color='orange')

# Add vertical lines and annotations for periods
for boundary in interval_boundaries:
    plt.axvline(x=boundary, color='gray', linestyle='--', alpha=0.5)
    idx = int(boundary / delta_t)
    plt.text(boundary + 0.1, np.min(true_values_cumulative) * 1.05, 
             f'σ_P={sigma_P_values[idx]:.2f}\nσ_n={sigma_n_values[idx]:.2f}', 
             horizontalalignment='left', fontsize=8, color='black')

plt.xlabel('Time (hours)')
plt.ylabel('Cumulative Value')
plt.title('Cumulative Market Price and True Value')
plt.legend()
plt.grid(True)
plt.show()

# Plot cumulative insider profits
cumulative_insider_profits = np.cumsum(insider_profits)
plt.figure(figsize=(12, 6))
plt.plot(times, cumulative_insider_profits, label='Cumulative Insider Profit', color='green')
plt.xlabel('Time (hours)')
plt.ylabel('Profit')
plt.title('Cumulative Insider Profit')
plt.legend()
plt.grid(True)
plt.show()

# Plot lambda evolution
plt.figure(figsize=(12, 6))
plt.plot(times, lambda_values, label='Lambda over Time', color='purple')
plt.xlabel('Time (hours)')
plt.ylabel('Lambda')
plt.title('Evolution of Lambda')
plt.legend()
plt.grid(True)
plt.show()
```

## Simplified Approach: Calculating $\lambda$ Using Covariance

Before introducing the optimization-based method, we used a simplified approach to compute $\lambda$, based on the relationship between the covariance of $P$ and $Y$, and the variance of $Y$. This approach is derived from the theoretical equilibrium condition in the Kyle model:

$
\lambda = \frac{\text{Cov}(P, Y)}{\text{Var}(Y)}
$

### Explanation of the Method:

1. **Market Makers' Sensitivity ($\lambda$)**:
   - $\lambda$ measures how the market price reacts to observed order flow $Y$, where:
     $
     Y = \beta P + n
     $
   - Using the properties of covariance and variance:
     $
     \text{Var}(Y) = \beta^2 \sigma_P^2 + \sigma_n^2
     $
     $
     \text{Cov}(P, Y) = \beta \sigma_P^2
     $
   - Substituting these into the formula:
     $
     \lambda = \frac{\beta \sigma_P^2}{\beta^2 \sigma_P^2 + \sigma_n^2}
     $

2. **Application**:
   - This formula assumes $\beta$ is pre-determined (e.g., $\beta = \frac{\sigma_n}{\sigma_P}$) and allows for a quick estimation of $\lambda$.
   - It is particularly useful for large sample windows where $\text{Cov}(P, Y)$ and $\text{Var}(Y)$ can be reliably estimated.

3. **Limitations**:
   - This method does not explicitly enforce the market makers’ zero-expected-profit condition.
   - It provides an approximation but may not align perfectly with the theoretical equilibrium when $\beta$ deviates from the optimal value.

### Why Include This Method?

This covariance-based approach is a simplified, computationally efficient way to estimate $\lambda$, particularly in scenarios where optimization frameworks are unavailable. While less precise than the optimization-based method, it offers valuable insights and serves as a useful baseline for comparison.


```python
import numpy as np
import matplotlib.pyplot as plt
from collections import namedtuple

# Définir la structure TimeSlice
TimeSlice = namedtuple('TimeSlice', ['t', 'P', 'n'])

# Paramètres de simulation
total_hours = 24     # Durée totale de la simulation en heures
delta_t = 0.1        # Pas de temps en heures (6 minutes)
times = np.arange(0, total_hours + delta_t, delta_t)  # Liste des temps

# Durée des intervalles de 3 heures
interval_length = 3
interval_points = int(interval_length / delta_t)

# Définir les sigmas pour chaque temps
sigma_P_values = []
sigma_n_values = []
interval_boundaries = []  # To store the interval boundaries for plotting

for i, t in enumerate(times):
    # Enregistrer les limites des intervalles de 3 heures
    if i % interval_points == 0:
        interval_boundaries.append(t)

    # Pendant l'événement (par exemple, de la 12ème à la 15ème heure)
    if 12 <= t < 15:
        sigma_P = 2.0  # Volatilité élevée de P
        sigma_n = 0.5  # Volatilité faible des noise traders
    else:
        sigma_P = np.random.uniform(0.1, 0.5)  # Volatilité faible pour P
        sigma_n = np.random.uniform(1.5, 3.0)  # Volatilité élevée pour noise traders
    
    sigma_P_values.append(sigma_P)
    sigma_n_values.append(sigma_n)

# Vérifier que les listes ont la bonne longueur
assert len(sigma_P_values) == len(times)
assert len(sigma_n_values) == len(times)

# Générer les valeurs de P et n pour chaque temps
P_values = []
n_values = []

np.random.seed(42)  # Pour la reproductibilité

for sigma_P, sigma_n in zip(sigma_P_values, sigma_n_values):
    P = np.random.normal(0, sigma_P * np.sqrt(delta_t))
    n = np.random.normal(0, sigma_n * np.sqrt(delta_t))
    P_values.append(P)
    n_values.append(n)

# Créer la liste des TimeSlice
Sim = [TimeSlice(t=t, P=P, n=n) for t, P, n in zip(times, P_values, n_values)]

# Calculer la taille de la fenêtre glissante
window_duration = 4  # Durée de la fenêtre en heures
window_size = int(window_duration / delta_t)  # Taille de la fenêtre en nombre de points

# Initialiser les listes pour stocker les valeurs de lambda, h(Y), V, Y et profit de l'insider
lambda_values = []
h_Y_values = []
V_values = []
Y_values = []
insider_profits = []

for i in range(len(Sim)):
    # Obtenir le TimeSlice actuel
    ts = Sim[i]
    
    # Définir la fenêtre
    start_index = max(0, i - window_size + 1)
    window = Sim[start_index:i+1]
    
    # Extraire les valeurs de P et n dans la fenêtre
    P_window = np.array([tsw.P for tsw in window])
    n_window = np.array([tsw.n for tsw in window])
    
    # Calculer le beta pour chaque TimeSlice dans la fenêtre en fonction de sigma_P et sigma_n
    beta_window = np.array([sigma_n_values[start_index + j] / sigma_P_values[start_index + j] for j in range(len(window))])
    V_window = beta_window * P_window  # Utiliser beta variable dans la fenêtre
    
    # Calculer V pour le temps actuel
    beta_current = sigma_n_values[i] / sigma_P_values[i]  # Beta correspondant au temps actuel
    V_current = beta_current * ts.P
    V_values.append(V_current)
    
    # Calculer Y dans la fenêtre
    Y_window = V_window + n_window
    Y_current = V_current + ts.n
    Y_values.append(Y_current)
    
    # Calculer lambda du market maker
    if len(Y_window) > 1:
        cov_PY = np.cov(P_window, Y_window, bias=True)[0, 1]
        var_Y = np.var(Y_window)
        lambda_mm = cov_PY / var_Y if var_Y != 0 else 0.0
    else:
        lambda_mm = 0.0  # Valeur initiale
    
    lambda_values.append(lambda_mm)
    
    # Calculer h(Y) pour le temps actuel
    h_Y = lambda_mm * Y_current
    h_Y_values.append(h_Y)
    
    # Calculer le profit de l'insider: (P - h(Y)) * V
    insider_profit = (ts.P - h_Y) * V_current
    insider_profits.append(insider_profit)

# Calculer les prix cumulés
market_prices_cumulative = np.cumsum(h_Y_values)
true_values_cumulative = np.cumsum([ts.P for ts in Sim])

# Calculer le profit cumulé de l'insider
cumulative_insider_profits = np.cumsum(insider_profits)

# Tracer les prix cumulés
plt.figure(figsize=(12, 6))
plt.plot(times, market_prices_cumulative, label='Prix du Market Maker h(Y)')
plt.plot(times, true_values_cumulative, label='Valeur Réelle de l\'Actif P')

# Ajouter des lignes verticales pour chaque changement d'intervalle de 3 heures
for boundary in interval_boundaries:
    plt.axvline(x=boundary, color='gray', linestyle='--', alpha=0.5)
    idx = int(boundary / delta_t)  # Index du point de la borne
    
    # Ajouter les textes en bas du graphique, avec saut de ligne entre σ_P et σ_n
    plt.text(boundary + 0.1, np.min(true_values_cumulative) * 1.07, 
             f'σ_P={sigma_P_values[idx]:.2f}\nσ_n={sigma_n_values[idx]:.2f}', 
             horizontalalignment='left', fontsize=8, color='black')

plt.xlabel('Temps (heures)')
plt.ylabel('Prix Cumulé')
plt.title('Évolution du Prix de l\'Actif (Cumulé)')
plt.legend()
plt.grid(True)
plt.show()

# Visualiser l'évolution du profit cumulé de l'insider
plt.figure(figsize=(12, 6))
plt.plot(times, cumulative_insider_profits, label='Profit Cumulé de l\'Insider', color='green')
plt.xlabel('Temps (heures)')
plt.ylabel('Profit Cumulé')
plt.title('Évolution du Profit Cumulé de l\'Insider')
plt.legend()
plt.grid(True)
plt.show()

# Visualiser l'évolution de lambda
plt.figure(figsize=(12, 6))
plt.plot(times, lambda_values, label='Lambda du Market Maker')
plt.xlabel('Temps (heures)')
plt.ylabel('Lambda')
plt.title('Évolution de Lambda du Market Maker')
plt.legend()
plt.grid(True)
plt.show()
```

# Modeling Market Dynamics Using a Two-State Markov Chain

---

## **1. Objective**

The goal is to simulate market price dynamics using a **two-state Markov chain**, where the system alternates between states that represent different market conditions. This approach models how market parameters, such as price sensitivity ($\lambda$) and asset volatility ($\sigma_P^2$), evolve dynamically.

---

## **2. Markov Chain Dynamics**

### 2.1 States

1. **State 0 (Inferior)**:
   - Represents a period of market consolidation dominated by noise traders.
   - **High probability of remaining in this state**: $ P(\text{Inferior} \to \text{Inferior}) = 0.95 $.

2. **State 1 (Superior)**:
   - Represents a period of high insider activity triggered by new information.
   - **Low probability of occurrence**: $ P(\text{Inferior} \to \text{Superior}) = 0.05 $.

### 2.2 Transition Matrix

The transition matrix for the Markov chain is defined as:

$
T = \begin{bmatrix}
0.95 & 0.05 \\
0.80 & 0.20
\end{bmatrix}
$

where $ T_{ij} $ represents the probability of transitioning from state $ i $ to state $ j $.


---

## **3. Market Parameters**

### 3.1 Price Sensitivity ($\lambda$)
The sensitivity of market prices to order flow is modeled as:

$
\lambda = \lambda_0 + \lambda_1 \cdot u
$

where:
- $ u \in \{0, 1\} $ is the current Markov state.
- $ \lambda_0 $: Baseline sensitivity.
- $ \lambda_1 $: Incremental sensitivity in the Superior state.

### 3.2 Asset Variance ($\sigma_P^2$)
The variance of the asset price is modeled as:

$
\sigma_P^2 = \sigma_0^2 + \sigma_1^2 \cdot u
$

where:
- $ \sigma_0^2 $: Baseline variance.
- $ \sigma_1^2 $: Incremental variance in the Superior state.

---

## **4. Simulation Workflow**

### 4.1 Simulating the Markov Chain
- Start with an initial state ($ u_0 = 0 $).
- Simulate state transitions over the total simulation duration using the transition matrix.

### 4.2 Generating Parameters
At each time step $ t $:
- Compute $\lambda_t$ and $\sigma_{P, t}^2$ based on the current Markov state $ u_t $.

### 4.3 Generating Asset Prices
- Simulate the real asset value $ P_t $ as a Gaussian random variable:
  $
  P_t \sim \mathcal{N}(0, \sigma_{P, t}^2)
  $
- Simulate the noise trader volume $ n_t $ as:
  $
  n_t \sim \mathcal{N}(0, 1)
  $

### 4.4 Market Maker's Pricing
- Calculate the total order flow:
  $
  Y_t = \lambda_t (P_t + n_t)
  $
- Market price is a noisy observation of $ Y_t $:
  $
  h(Y_t) \sim \mathcal{N}(Y_t, \epsilon^2)
  $
  where $\epsilon$ represents pricing noise.


```python
import numpy as np
import matplotlib.pyplot as plt

# Simulation Parameters
total_hours = 24
delta_t = 0.1  # Time step (hours)
steps = int(total_hours / delta_t)

# Markov Chain Transition Matrix
transition_matrix = np.array([[0.95, 0.05],  # Inférieur -> Inférieur, Supérieur
                               [0.80, 0.20]])  # Supérieur -> Inférieur, Supérieur

# Initial State (0: Inférieur, 1: Supérieur)
state = 0
states = [state]

# Simulate Markov Chain States
for _ in range(steps):
    state = np.random.choice([0, 1], p=transition_matrix[state])
    states.append(state)

# Parameters for λ and σ² based on state
lambda_0, lambda_1 = 0.1, 0.5
sigma_0_sq, sigma_1_sq = 0.01, 0.04

lambda_values = [lambda_0 + lambda_1 * state for state in states]
sigma_P_sq_values = [sigma_0_sq + sigma_1_sq * state for state in states]

# Generate P and n values
P_values = [np.random.normal(0, np.sqrt(sigma)) for sigma in sigma_P_sq_values]
n_values = np.random.normal(0, 1, len(P_values))  # Noise traders' orders

# Market Maker's Observations
Y_values = [lambda_ * (P + n) for lambda_, P, n in zip(lambda_values, P_values, n_values)]
h_Y_values = [np.random.normal(Y, 0.1) for Y in Y_values]  # Noisy market prices

# Plot State Dynamics
plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(states)) * delta_t, states, label='Markov State (0: Inférieur, 1: Supérieur)', color='purple')
plt.xlabel('Time (hours)')
plt.ylabel('State')
plt.title('State Dynamics (Markov Chain)')
plt.legend()
plt.grid(True)
plt.show()

# Plot λ and σ² Dynamics
plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(lambda_values)) * delta_t, lambda_values, label=r'$\lambda$ Values', color='blue')
plt.plot(np.arange(len(sigma_P_sq_values)) * delta_t, sigma_P_sq_values, label=r'$\sigma_P^2$ Values', color='orange')
plt.xlabel('Time (hours)')
plt.ylabel('Parameter Values')
plt.title('Dynamics of $\lambda$ and $\sigma_P^2$')
plt.legend()
plt.grid(True)
plt.show()

# Plot Market Prices
plt.figure(figsize=(12, 6))
plt.plot(np.arange(len(h_Y_values)) * delta_t, h_Y_values, label='Market Price h(Y)', color='green')
plt.xlabel('Time (hours)')
plt.ylabel('Market Price')
plt.title('Market Price Dynamics')
plt.legend()
plt.grid(True)
plt.show()
```

# Simulating Market Prices with Different Distributions

---

## Objective

We now explore the impact of using different distributions on the price simulation in the Kyle model. Specifically, we use:
1. The **Laplace Distribution**: Characterized by a sharp peak and heavier tails compared to the normal distribution.
2. The **Beta Distribution**: Bounded within [0, 1], ideal for modeling relative or proportional values.

These simulations aim to understand how changes in the distributions of $P$ (asset value) and $n$ (noise traders) affect price sensitivity ($\lambda$) and the insider's strategy ($\beta$).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize
from collections import namedtuple

# Simulation function
def simulate_kyle_model(P_values, n_values, sigma_P_values, sigma_n_values, times, delta_t):
    """
    Runs the Kyle model simulation for given P and n values.

    Parameters:
        P_values (np.ndarray): Array of P values (asset values).
        n_values (np.ndarray): Array of n values (noise trader volumes).
        sigma_P_values (np.ndarray): Dynamic standard deviations for P.
        sigma_n_values (np.ndarray): Dynamic standard deviations for n.
        times (np.ndarray): Array of time steps.
        delta_t (float): Time step size.

    Returns:
        dict: Simulation results including lambda, prices, and insider profits.
    """
    # Define the sliding window duration
    window_duration = 4  # 4 hours
    window_size = int(window_duration / delta_t)

    lambda_values = []
    insider_profits = []
    market_prices_cumulative = []
    true_values_cumulative = []

    # Function to solve lambda given beta
    def solve_lambda(beta, P_window, n_window):
        def objective(lambda_):
            Y_window = beta * P_window + n_window
            h_Y = lambda_ * Y_window
            return -np.mean((P_window - h_Y) * (beta * P_window))  # Insider gain (negative for maximization)

        def constraint(lambda_):
            Y_window = beta * P_window + n_window
            h_Y = lambda_ * Y_window
            return np.mean((h_Y - P_window) * Y_window)  # Market maker zero-profit constraint

        result = minimize(objective, x0=0.1, constraints={'type': 'eq', 'fun': constraint}, bounds=[(0, None)])
        return result.x[0] if result.success else 0.0

    for i in range(len(times)):
        # Define the sliding window
        start_idx = max(0, i - window_size + 1)
        P_window = P_values[start_idx:i+1]
        n_window = n_values[start_idx:i+1]

        # Solve for beta and lambda dynamically
        sigma_P = sigma_P_values[i]
        sigma_n = sigma_n_values[i]
        beta = sigma_n / sigma_P  # Theoretical beta
        lambda_mm = solve_lambda(beta, P_window, n_window)

        # Calculate V, Y, and h(Y) for the current time step
        V_current = beta * P_values[i]
        Y_current = V_current + n_values[i]
        h_Y_current = lambda_mm * Y_current

        # Calculate profits
        insider_profit = (P_values[i] - h_Y_current) * V_current
        insider_profits.append(insider_profit)

        # Track cumulative values
        if i == 0:
            market_prices_cumulative.append(h_Y_current)
            true_values_cumulative.append(P_values[i])
        else:
            market_prices_cumulative.append(market_prices_cumulative[-1] + h_Y_current)
            true_values_cumulative.append(true_values_cumulative[-1] + P_values[i])

        lambda_values.append(lambda_mm)

    return {
        "lambda_values": lambda_values,
        "market_prices_cumulative": market_prices_cumulative,
        "true_values_cumulative": true_values_cumulative,
        "insider_profits": np.cumsum(insider_profits),
    }

# Parameters for the simulation
total_hours = 24
delta_t = 0.1
times = np.arange(0, total_hours + delta_t, delta_t)
interval_length = 3
interval_points = int(interval_length / delta_t)

# Generate dynamic sigma_P and sigma_n values
sigma_P_values = []
sigma_n_values = []
for t in times:
    if 12 <= t < 15:  # Shock period
        sigma_P = 2.0
        sigma_n = 0.5
    else:  # Consolidation period
        sigma_P = np.random.uniform(0.1, 0.5)
        sigma_n = np.random.uniform(1.5, 3.0)
    sigma_P_values.append(sigma_P)
    sigma_n_values.append(sigma_n)
```
